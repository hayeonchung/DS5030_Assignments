- Imagine rolling a fair single six-sided die. There are 6 outcomes, all equally likely. Derive the sample space and the space of events. What are the probabilities of the outcomes and events? 
- Consider a random variable that assigns the square root of the number of pips on the die to each outcome. Write code to simulate rolling a single six-sided die and computing the value of the random variable. Simulate 5000 rolls and plot the mass function and ECDF of the random variable.
- Imagine rolling two fair six-sided die. Consider a random variable that adds up the pips on the dice. There are 11 outcomes (2 , 3, ..., 12), but not all are equally likely. Derive the sample space and **describe** the space of events. What are the probabilities of the outcomes?

a) The sample space S={1,2,3,4,5,6}. The space of events P(S)=all subsets of S which is 2^6=64 events. The probabilities of the outcomes and events are uniform, where for any A in S, P(A) = |A|/6. 

```python
# b)
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(53)      
n_rolls = 5000          
die_faces = np.arange(1, 7)  # possible die outcomes: 1–6

# Simulate rolling a single six-sided die
rolls = np.random.choice(die_faces, size=n_rolls, replace=True)

# Define the random variable Y = sqrt(X)
Y = np.sqrt(rolls)

# Compute empirical PMF
unique_vals, counts = np.unique(Y, return_counts=True)
pmf = counts / n_rolls

# Plot the empirical PMF
plt.figure(figsize=(7,4))
plt.bar(unique_vals, pmf, width=0.1, color='blue', edgecolor='black')
plt.title("Empirical PMF of Y = √(X) for a Single Die (5000 Rolls)")
plt.xlabel("Y = √(X)")
plt.ylabel("Probability")
plt.show()

# Compute and plot ECDF
sorted_Y = np.sort(Y)
ecdf = np.arange(1, len(sorted_Y) + 1) / len(sorted_Y)

plt.figure(figsize=(7,4))
plt.step(sorted_Y, ecdf, where='post', color='blue')
plt.title("Empirical CDF of Y = √(X) for a Single Die (5000 Rolls)")
plt.xlabel("Y = √(X)")
plt.ylabel("ECDF")
plt.show()
```

The event space is the complete collection of all possible outcomes or sets of outcomes that could occur when both dice are rolled. Each individual outcome is an ordered pair showing the number on the first die and the number on the second, such as (2, 5) or (6, 3). Together, these 36 equally likely outcomes form the sample space. The event space then consists of every possible combination or grouping of these outcomes (all subsets of the sample space). Each event corresponds to some statement about the dice, like “the sum is even,” “the total is greater than 9,” or “both dice show the same number.” Because each event is a subset of the 36 possible ordered pairs, the event space is the set of all these subsets and probabilities can be assigned to events by adding up the probabilities of their individual outcomes.

```python
# d)
# Simulation parameters
np.random.seed(53)       
n_rolls = 10000          
die_faces = np.arange(1, 7)  # sides of a six-sided die

# Simulate rolling two dice
die1 = np.random.choice(die_faces, size=n_rolls, replace=True)
die2 = np.random.choice(die_faces, size=n_rolls, replace=True)

# Random variable S = sum of both dice
S = die1 + die2

# Compute empirical PMF
unique_sums, counts = np.unique(S, return_counts=True)
pmf = counts / n_rolls

# Plot PMF
plt.figure(figsize=(7,4))
plt.bar(unique_sums, pmf, color='blue', edgecolor='black')
plt.title("Empirical PMF of S = Die1 + Die2 (10,000 Rolls)")
plt.xlabel("Sum of Two Dice (S)")
plt.ylabel("Probability")
plt.grid(alpha=0.3)
plt.show()

# Compute and plot ECDF
sorted_S = np.sort(S)
ecdf = np.arange(1, len(sorted_S) + 1) / len(sorted_S)

plt.figure(figsize=(7,4))
plt.step(sorted_S, ecdf, where='post', color='blue')
plt.title("Empirical CDF of S = Die1 + Die2 (10,000 Rolls)")
plt.xlabel("Sum of Two Dice (S)")
plt.ylabel("ECDF")
plt.grid(alpha=0.3)
plt.show()
```

### 4. Roulette

Roulette is a betting game. There are 37 possible outcomes: A green 0, and the numbers 1 to 36 in red and black. Here is a picture of the (American, not Euro, it has an extra green 00) betting board:

![Roulette](./src/euro_roulette.jpg)

To bet, you must pay a dollar, but then you get payouts that depend on how many slots are in your bet
- Basic bets:
    - Red or Black slots
    - Odd or Even slots
    - A single slots, like 20
- More complex bets:
    - Split: Two adjacent slots (e.g. {1,2})
    - Square: Four adjacent slots (e.g. {1,2,4,5})
    - Street: Three slots in a row (e.g. {1,2,3})
    - Line: Six slots (e.g. {1,2,3,4,5,6})
In general, you can only bet on 1, 2, 3, 4, 6 slots, 12 slots, or 18 slots. If your bet occurs when the wheel is spun, you gain 36/K-1 where $K$ is the number of slots you bet on; if not, you lose a dollar and get -1.

- Write code to model spinning the roulette wheel, including the colors and numbers (you could make two lists of number and color and draw a random number between 0 and 37... or use a dataframe with color and number variables and sample it... or use a dict with key to number/color pairs...)
- Describe the probability space associated with the roulette wheel: Outcomes, events, probabilities (If there are 37 outcomes, there are $2^{37}= 137,438,953,472$ events, by the way)
- You wrote code to generate a spin of the roulette wheel. Now write a function that takes a basic or complex bet as an argument, and returns the result for the player (win or lose, and the payout 36/K-1 or -1)
- Simulate betting on red, betting on odd, betting on 7, a split, and a line 1000 times each.
- Compute the average values for the bets you just simulated. What are the expected average payoffs?

```python
import random
from statistics import mean
random.seed(53)

# 1) Wheel model: single-zero (37 outcomes)
GREEN = {0}
RED = {1,3,5,7,9,12,14,16,18,19,21,23,25,27,30,32,34,36}
BLACK = set(range(1,37)) - RED
ALL_NUMBERS = {0} | set(range(1,37))  # 0..36

def spin():
    """Return the outcome of one roulette spin (an int in {0..36})."""
    return random.randrange(0, 37)  # uniform over 37 outcomes

# 2) Generic bet resolver 
def resolve_bet(outcome, bet):
    """
    Resolve a bet against an outcome.
    bet is a dict with one of:
      - {"type": "color", "color": "red"|"black"}
      - {"type": "parity", "parity": "odd"|"even"}  # 0 is neither
      - {"type": "numbers", "nums": {…}}            # any set of K numbers
    Returns net payoff using rule: win: 36/K - 1, lose: -1
    """
    if bet["type"] == "color":
        if bet["color"] == "red":
            covered = RED
        else:
            covered = BLACK
        K = len(covered)                 # K = 18
        win = outcome in covered

    elif bet["type"] == "parity":
        if bet["parity"] == "odd":
            covered = {n for n in range(1,37) if n % 2 == 1}
        else:
            covered = {n for n in range(1,37) if n % 2 == 0}
        K = len(covered)                 # K = 18
        win = outcome in covered

    elif bet["type"] == "numbers":
        covered = set(bet["nums"])
        # Safety checks
        assert covered.issubset(ALL_NUMBERS), "Bet includes invalid numbers"
        assert len(covered) >= 1, "K must be >= 1"
        K = len(covered)
        win = outcome in covered

    else:
        raise ValueError("Unknown bet type")

    return (36 / K - 1.0) if win else -1.0

# 3) Helper: simulate an average payoff 
def simulate_avg_payoff(bet, trials=1000):
    results = [resolve_bet(spin(), bet) for _ in range(trials)]
    return mean(results)

# 4) Theoretical expected value for any K-number bet 
def theoretical_ev(K, n_outcomes=37):
    # p_win = K/n; payoff if win = 36/K - 1; else -1
    p = K / n_outcomes
    return p * (36 / K - 1) + (1 - p) * (-1)

# 5) Define bets and run 
bets = {
    "Red":           {"type": "color", "color": "red"},                       # K = 18
    "Odd":           {"type": "parity", "parity": "odd"},                     # K = 18
    "Single 7":      {"type": "numbers", "nums": {7}},                        # K = 1
    "Split {1,2}":   {"type": "numbers", "nums": {1, 2}},                     # K = 2
    "Line {1..6}":   {"type": "numbers", "nums": {1,2,3,4,5,6}},              # K = 6
}

TRIALS = 1000

print("Wheel: single-zero (37 outcomes). Payoff rule: win: 36/K - 1, lose: -1\n")
for name, bet in bets.items():
    avg = simulate_avg_payoff(bet, trials=TRIALS)

    # Determine K for theory
    if   bet["type"] == "color":   K = 18
    elif bet["type"] == "parity":  K = 18
    else:                          K = len(bet["nums"])

    ev = theoretical_ev(K)
    print(f"{name:14s}  K={K:2d}  Empirical avg over {TRIALS:4d} spins: {avg: .5f}   Theoretical EV: {ev: .5f}")
```

The probability space for this roulette wheel consists of three parts: the sample space, the event space, and the probability function. The sample space includes all possible outcomes when the wheel is spun which is the 37 numbered slots on the wheel: 0 (green) and the integers 1-36 alternating in red and black. Each of these outcomes is equally likely, so each has a probability of 1/37.

The event space is the set of all possible events that can occur, where each event is any subset of the 37 possible outcomes. For example, one event could be “the ball lands on red,” which includes 18 outcomes, or “the ball lands on an even number,” which also includes 18 outcomes. Since there are 37 individual outcomes, the total number of possible events is 2^37. The probability function assigns each individual outcome a probability of 1/37 and computes the probability of an event by summing the probabilities of the outcomes it contains. Thus, the probability space can be described as 
(Ω,F,P), where Ω is the set of 37 outcomes, F is the collection of all 
2^37 subsets of Ω,and P gives each single outcome an equal probability of 1/37.

- Logistic distribution (similar to normal): 
$$
F(x; \sigma) = \dfrac{1}{1+e^{-x/\sigma}} 
$$
with $\sigma >0$.
- Exponential distribution (similar to log-normal):
$$
F_X(x) = \begin{cases}
0, & x<0 \\
1 - e^{-\lambda x}, & x \ge 0,
\end{cases}
$$
with $\lambda > 0$.
- Negative Binomial (similar to Poisson): The probability mass function for positive integers is:
$$
f(k;r,p) = \dfrac{(k+r-1)!}{k!(r-1)!}(1-p)^k p^r, \quad \text{ for $k=0,...n$}
$$
You can interpret this as follows: Flip a coin that comes up heads with probability $p$ until you get $r$ heads, and then stop. What is the probability of stopping at each $k=0,1,2,...$?
- Categorical (similar to Bernoulli): The probability mass function over $k = 1, 2, ..., K$ categories is
$$
f(k;p_1,...,p_K) = p_1^{k=1}p_2^{k=2}...p_K^{k=K}
$$
where $0 \le p_i \le 1$ and $ \sum_{k=1}^K p_k = 1$.

### 4) Categorical Distribution

a) Support: x ∈ {c_1, c_2, c_3,...} where each c_i represents a possible category within the variable. 

```python
# c) 
parameter_sets = [
    [0.10, 0.30, 0.40, 0.20],
    [0.25, 0.25, 0.25, 0.25],
    [0.05, 0.15, 0.60, 0.20],
]

K = 4
cats = np.arange(1, K+1)
n = 1000

plt.figure(figsize=(14, 6))

for i, p in enumerate(parameter_sets, 1):
    p = np.array(p)
    samples = np.random.choice(cats, size=n, p=p)

    # Empirical PMF (integer-aligned)
    counts = np.bincount(samples, minlength=K+1)[1:] / n
    cdf_theory = np.cumsum(p)
    cdf_emp = np.cumsum(counts)

    # PMF panel
    ax1 = plt.subplot(2, len(parameter_sets), i)
    ax1.bar(cats, counts, width=0.8, alpha=0.7, label="Empirical (n=1000)")
    ax1.plot(cats, p, "r--", label="Theoretical PMF")
    ax1.set(
        title=f"Categorical PMF, p={p.tolist()}",
        xlabel="Category", ylabel="Probability",
        xlim=(0.5, K+0.5), ylim=(0, max(0.35, 1.1*max(p)))
    )
    ax1.legend()

    # CDF panel (right-continuous step)
    ax2 = plt.subplot(2, len(parameter_sets), i + len(parameter_sets))
    ax2.step(cats, cdf_emp, where="post", label="Empirical ECDF")
    ax2.step(cats, cdf_theory, where="post", linestyle="--", label="Theoretical CDF", color="r")
    ax2.set(title="Empirical vs Theoretical CDF",
            xlabel="Category", ylabel="Cumulative Probability",
            xlim=(0.5, K+0.5), ylim=(0, 1.02))
    ax2.legend()

plt.tight_layout()
plt.show()
```

```python
# d) 
x = df['Tumor Stage'].dropna()
counts = x.value_counts(normalize=True)
categories = counts.index
pmf = counts.values
cdf = np.cumsum(pmf)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
sns.barplot(x=categories, y=pmf, color='lightblue')
plt.title("Categorical PMF: Tumor Stage")
plt.ylabel("Probability")

plt.subplot(1,2,2)
plt.step(categories, cdf, where="post", color='r')
plt.title("Categorical CDF: Tumor Stage")
plt.ylabel("Cumulative Probability")
plt.tight_layout()
plt.show()
```
